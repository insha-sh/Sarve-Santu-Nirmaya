# Data Pre-processing 

## Importing libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Importing the dataset

```python
dataset = pd.read_excel('Healthcare_Workers.xlsx', 'Sheet2')
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset = dataset.astype({'CHILDREN': 'object',
                          'ELDERS' : 'object',
                          'CHRONIC' : 'object',
                          'MEALS_SKIP' : 'object', 
                          'DEATHS_FAMILY' :'object',
                          'REASON_FAM_DEATH_COVID' : 'object',
                          'MENTAL_HEALTH_MEDS_BEFORE_COVID' : 'object',
                         'DEATHS_WORK' : 'object'})
```

```python
dataset.info()
```

```python
dataset.describe()
```

## Exploratory Data Analysis

```python
sns.set_theme(style = "darkgrid", rc= {"axes.facecolor": ".3",'grid.linewidth': 0.1, })

plt.figure(figsize = (20,10))

plt.subplot(2,4,1)
sns.histplot(data = dataset, x = 'AGE', color = 'White', stat = 'density', kde = True)

plt.subplot(2,4,2)
sns.histplot(data = dataset, x = 'TRAVEL_MIN', color = 'White', stat = 'density', kde = True,
            binwidth = 30)

plt.subplot(2,4,3)
sns.histplot(data = dataset, x = 'WORK_HOURS', color = 'White', stat = 'density', kde = True)

plt.subplot(2,4,4)
sns.histplot(data = dataset, x = 'HOLIDAY_MONTH', color = 'White', stat = 'density', kde = True)

plt.subplot(2,4,5)
sns.histplot(data = dataset, x = 'SLEEP_HOURS_24', color = 'White', stat = 'density', kde = True)

plt.subplot(2,4,6)
sns.histplot(data = dataset, x = 'FEAR_SCORE', color = 'White', stat = 'density', kde = True)

plt.subplot(2,4,7)
sns.histplot(data = dataset, x = 'ANXIETY_SCORE', color = 'White', stat = 'density', kde = True)

plt.subplot(2,4,8)
sns.histplot(data = dataset, x = 'PTSD_SCORE', color = 'White', stat = 'density', kde = True)
```

```python
# Advanced statistical methods
grid = sns.PairGrid(data = dataset[['AGE',
                          'TRAVEL_MIN',
                          'WORK_HOURS',
                          'HOLIDAY_MONTH',
                          'SLEEP_HOURS_24',
                          'FEAR_SCORE',
                          'ANXIETY_SCORE',
                          'PTSD_SCORE']], )
grid.map(sns.scatterplot)
```

```python
plt.figure(figsize = (10,7))

sns.heatmap(dataset.corr(method = "kendall"), annot = True, cmap='magma')
```

```python
#Alone and nuclear

alone_dataset = dataset[dataset['FAM_TYPE'] == 'Alone']
nuclear_dataset = dataset[dataset['FAM_TYPE'] == 'Nuclear']

plt.figure(figsize = (30,10))

plt.subplot(1,2,1)
sns.heatmap(alone_dataset.corr(method = "kendall"), annot = True, cmap="magma")
plt.title('FAMILY TYPE - ALONE')

plt.subplot(1,2,2)
sns.heatmap(nuclear_dataset.corr(method = "kendall"), annot = True,cmap="magma")
plt.title("FAMILY TYPE - NUCLEAR")
```

```python
#Male and Female

male_dataset = dataset[dataset['GENDER'] == 'Male']
female_dataset = dataset[dataset['GENDER'] == 'Female']

plt.figure(figsize = (30,10))

plt.subplot(1,2,1)
sns.heatmap(male_dataset.corr(method = "kendall"), annot = True, cmap="magma")
plt.title('GENDER - MALE')

plt.subplot(1,2,2)
sns.heatmap(female_dataset.corr(method = "kendall"), annot = True,cmap="magma")
plt.title("GENDER - FEMALE")
```

```python
#Family with kids vs without

childno_dataset = dataset[dataset['CHILDREN'] == 0]
childyes_dataset = dataset[dataset['CHILDREN'] == 1]

plt.figure(figsize = (30,10))

plt.subplot(1,2,1)
sns.heatmap(childno_dataset.corr(method = "kendall"), annot = True, cmap="magma")
plt.title('Individuals without kids at home')

plt.subplot(1,2,2)
sns.heatmap(childyes_dataset.corr(method = "kendall"), annot = True,cmap="magma")
plt.title("Individuals with kids at home")
```

```python
#Family with kids vs without

chronicno_dataset = dataset[dataset['CHRONIC'] == 0]
chronicyes_dataset = dataset[dataset['CHRONIC'] == 1]

plt.figure(figsize = (30,10))

plt.subplot(1,2,1)
sns.heatmap(chronicno_dataset.corr(method = "kendall"), annot = True, cmap="magma")
plt.title('Individuals without Chronic illness family members at home')

plt.subplot(1,2,2)
sns.heatmap(chronicyes_dataset.corr(method = "kendall"), annot = True,cmap="magma")
plt.title('Individuals with Chronic illness family members at home')
```
